```python
import ROOT as r 
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
directory= "/Users/adit/Google Drive/Internship_ Research/Muon g-2/Data Filed/combined_forAdit_timealigned.root"
f = r.TFile(directory)
f.cd('LaserEvents')
t = f.Get('LaserEvents/tree')
df = pd.DataFrame(r.RDataFrame(t).AsNumpy())
```

```python
c = r.TCanvas()
calo= 1
xtal = 1
```

```python
df_filtered = df.loc[(df['sipmTime']-df['slmTime']) < 50]
df_xtal = df_filtered[(df_filtered['caloNum']==calo)& (df_filtered['xtalNum']==xtal)]
syncPulse = df_xtal.groupby(['runNum','subRunNum','fillNum']).min()
syncPulse.rename({'slmEnergy': 'syncEnergy'}, axis = 'columns',inplace=True )
df_xtal  = pd.merge(df_xtal,syncPulse['syncEnergy'], how = 'inner', on = ['runNum','subRunNum','fillNum'])
df_xtal['normalSlmEnergy'] = (df_xtal['slmEnergy']/df_xtal['syncEnergy'])

mean_temp = df_xtal['sipmEnergy']/df_xtal['normalSlmEnergy'].mean() 
```

```python
h1 = r.TH1D(f"h_{1}", "Local Monitor 1 Energy of Calo {} xtal {}; LM Energy; Frequency".format(calo,xtal), 200, 1000, 1700)

for y in df_xtal['lmPmtAmp1']:
    h1.Fill(y)

h1.Draw()
c.SetLogy()
c.Draw()
```

```python
h2 = r.TH1D(f"h_{2}", "Energy Ratio LM1-SLM of Calo {} xtal {}; LM/SLM Energy; Frequency".format(calo,xtal), 200, 0, 2)

for y in df_xtal['lmPmtAmp1']/df_xtal['normalSlmEnergy']/mean_temp:
    h2.Fill(y)

h2.Draw()
c.Draw()
```

```python
h3 = r.TH1D(f"h_{3}", "Energy Ratio LM1-Calo of Calo {} xtal {}; LM/sipm Energy; Frequency".format(calo,xtal), 200, -1, 3)

for y in df_xtal['lmPmtAmp1']/df_xtal['sipmEnergy']:
    h3.Fill(y)

h3.Draw()
c.Draw()
```

```python
h4 = r.TH1D(f"h_{4}", "Local Monitor 2 Energy of Calo {} xtal {}; LM Energy; Frequency".format(calo,xtal), 200, 400, 1700)

for y in df_xtal['lmPmtAmp2']:
    h4.Fill(y)

h4.Draw()
c.Draw()
```

```python
h5 = r.TH1D(f"h_{5}", "Energy Ratio LM2-SLM of Calo {} xtal {}; LM/SLM Energy; Frequency".format(calo,xtal), 200, 0, 2)

for y in df_xtal['lmPmtAmp2']/df_xtal['normalSlmEnergy']/mean_temp:
    h5.Fill(y)

h5.Draw()
c.Draw()
```

```python
h6 = r.TH1D(f"h_{6}", "Energy Ratio LM2-Calo of Calo {} xtal {}; LM/sipm Energy; Frequency".format(calo,xtal), 200, 0, 1)

for y in df_xtal['lmPmtAmp2']/df_xtal['sipmEnergy']:
    h6.Fill(y)

h6.Draw()
c.Draw()
```
